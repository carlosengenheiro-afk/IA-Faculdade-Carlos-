```python
# Sistema de Classificação de Conforto Térmico baseado em regras

def classificar_conforto(temperatura, umidade):

    if temperatura >= 22 and temperatura <= 26 and umidade >= 30 and umidade <= 60:
        return "Confortável"

    elif temperatura > 26 and umidade > 60:
        return "Muito Quente e Úmido"

    elif temperatura > 26:
        return "Quente"

    elif temperatura < 22:
        return "Frio"

    else:
        return "Moderado"


# Testando o sistema
temperatura = 28
umidade = 70

resultado = classificar_conforto(temperatura, umidade)

print("Temperatura:", temperatura)
print("Umidade:", umidade)
print("Classificação:", resultado)
```

```python
import pandas as pd

dados = {
    "temperatura": [18,20,22,24,26,28,30,32,21,23,25,27],
    "umidade": [40,45,50,55,60,65,70,75,35,48,58,62],
    "conforto": [
        "Frio",
        "Frio",
        "Confortável",
        "Confortável",
        "Confortável",
        "Quente",
        "Muito Quente e Úmido",
        "Muito Quente e Úmido",
        "Confortável",
        "Confortável",
        "Confortável",
        "Quente"
    ]
}

df = pd.DataFrame(dados)

df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score

# Separando dados
X = df[["temperatura", "umidade"]]
y = df["conforto"]

# Transformar texto em número
encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y)

# Separar treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.3)

# Criar modelo
modelo = DecisionTreeClassifier()

# Treinar
modelo.fit(X_train, y_train)

# Testar
previsoes = modelo.predict(X_test)

# Acurácia
acuracia = accuracy_score(y_test, previsoes)

print("Acurácia do modelo:", acuracia)
```

```python
# Teste do modelo com novos dados

nova_temperatura = 29
nova_umidade = 68

previsao = modelo.predict([[nova_temperatura, nova_umidade]])

resultado = encoder.inverse_transform(previsao)

print("Temperatura:", nova_temperatura)
print("Umidade:", nova_umidade)
print("Previsão ML:", resultado[0])
```
